```python
## Modeling a Bank Account

class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self,amount):
        self.balance += amount
        print(f"{amount} is deposited. New Balance is {self.balance}")
        
    def withdraw(self, amount):
        if amount > self.balance:
            print(f"You do not have sufficient balance in your Account")
        else:
            self.balance -= amount
            print(f"{amount} has been deducted from your Bank Account. New Balance is {self.balance}")
    
    def get_balance(self):
        return self.balance
    
```

```python

#create an object
acc = BankAccount("Apoorv",5000)
acc.deposit(5000)
```

## INHERITENCE

```python
class Car:
    def __init__(self,windows,doors,engine_type):
        self.windows = windows
        self.doors = doors
        self.engine_type = engine_type

    def drive(self):
        print(f"The person will drive a {self.engine_type} car")
        
```

```python
car1 = Car(4,5,"Petrol")
car1.drive()
```

```python
class Tesla(Car):
    def __init__(self,windows,doors,engine_type,is_selfdrive):
        super().__init__(windows,doors,engine_type)
        self.is_selfdrive = is_selfdrive

    def self_drive(self):
        print(f"This Tesla has Self Drive? {self.is_selfdrive}")
```

```python
tesla1 = Tesla(4,5,"electric",True)
tesla1.self_drive()
tesla1.drive()
```

```python
## Multiple Inheritance
## when a class inherits from more than 1 class 

class Animal:
    def __init__(self,name):
        self.name = name

    def speak(self):
        print(f"Subclass must have this method")

class Pet:
    def __init__(self,owner):
        self.owner = owner


class Dog(Animal,Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)

    def speak(self):
        return f"{self.name} says woof!!!"
    
```

```python
dog = Dog("Puchku","Apoorv")
dog.speak()
```

## Polymorphism
 allows objects of different classes to be considered as objects of a common superclass. Way to perform single action in different forms. achieved with overriding and interfaces.

### Method Overriding
It allows a child class to provide a specific implementation of a method that is already defined in its parent class 

```python
## Base Class
class Animal:
    def speak(self):
        return "Sound of the animal"
    
class Dog(Animal):
    def speak(self):
        return "woof!!"
    
class Cat(Animal):
    def speak(self):
        return "Meow !!"
    
## function that demonstrates polymorphism
def animal_speak(animal):
    print(animal.speak())
```

```python
dog = Dog()
cat = Cat()
print(dog.speak(),cat.speak())
animal_speak(dog)
```

```python
## Methods

class Shape:
    def area(self):
        return "This returns the area of the shape"
    
class Rectangle(Shape):
    def __init__(self, width, length):
        self.width = width
        self.length = length

    def area(self):
        return self.width * self.length
    
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * (self.radius)**2
    
# function for polymorphism

def print_area(shape):
    print(f"The area is {shape.area()}")
```

```python
rect = Rectangle(4,5)
cir = Circle(3)
print_area(rect)
print_area(cir)
```

### polymorphism with Abstract Base Classes 
Used to define common methods for a group of related objects.





```python
from abc import ABC, abstractmethod

# define a abstract class
class Vehicle(ABC):
    @abstractmethod
    def start_eng(self):
        pass

#derived class
class Car(Vehicle):
    def start_eng(self):
        return "Engine Started: Car"
    
class Motorcycle(Vehicle):
    def start_eng(self):
        return "Engine Started: Bike"
    
def start_vehicle(vehicle):
    print(vehicle.start_eng())
    

## create objects
car = Car()
bike = Motorcycle()
start_vehicle(car)
start_vehicle(bike)
```

## Encapsulation And Abstraction 
Encapsulation is wrapping data and methods together as a single unit.it restricts direct access to some of the object's components.

Abstraction is hiding complex implementation details and exposing only necessary information


```python
## Encapsulation 
## Public Proctected Private

class Person:
    def __init__(self,name,age, gender):
        self.name = name # Public
        self.__age = age # Private Variables gets converted
        self._gender = gender # Protected Variable

class Employees(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)

def get_details(person):

    # ** YOU CAN'T CALL A PRIVATE VARIABLES
    #print(person.age)
    return person.name

empl = Employees("Apoorv", 24, "MALE")
print(empl._gender)
```

```python
## Encapsulation with Getter and Setter

class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

## Getter Method
    def get_name(self):
        return self.__name
    
    def get_age(self):
        return self.__age
    
## Setter Method
    def set_age(self,age):
        if age > 0:
            self.__age = age
        else:
            print("Age can't be negative")


person = Person("Apoorv",24)

print(person.get_name())
print(person.get_age())
person.set_age(25)
print(person.get_age())
person.set_age(-25)
```

## Abstraction 
hiding complex implementation details

```python
from abc import ABC, abstractmethod

## abstract classs

class Vehicle(ABC):
    def drive(self):
        print("The Vehicle is used for driving")

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car = Car()
operate_vehicle(car)
```

```python
## MAthematical operations for vectors 

class Vector:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __add__(self,other):
        return self.x+other.x, self.y+other.y
    def __sub__(self,other):
        return self.x-other.x,self.y-other.y
    def __mul__(self,other):
        return self.x*other.x, self.y*other.y
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
    

v1 = Vector(2,3)
v2 = Vector(4,5)

print(v1+v2)
print(v1-v2)
print(v1*v2)
print(v1==v2)
```
